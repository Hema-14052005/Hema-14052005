```python
from zipfile import ZipFile
import pandas as pd
data="/content/archive.zip"
with ZipFile(data,'r') as zp:
  zp.extractall()
d=pd.read_csv('/content/archive.zip')
d.head()
```

```python
#separating data
x=d.drop(columns='Outcome',axis=1)
y=d['Outcome']
```

```python
#data standardization
from sklearn.preprocessing import StandardScaler
x=d.drop(columns='Outcome',axis=1)
y=d['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
#training he model
from  sklearn import svm
model=svm.SVC(kernel="linear")
model.fit(x_test,y_test)
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
x_trainpre=model.predict(x_train)
x_testpre=model.predict(x_test)
train_acc=accuracy_score(x_trainpre,y_train)
testacc=accuracy_score(x_testpre,y_test)
print("training accuracy")
print(train_acc)
print('testing accuracy')
print(testacc)
```

```python
from zipfile import ZipFile
import numpy as np
from sklearn.preprocessing import StandardScaler
from  sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd

d=pd.read_csv('/content/diabetes.csv')
d.head()
x=d.drop(columns='Outcome',axis=1)
y=d['Outcome']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
model=svm.SVC(kernel="linear")
model.fit(x_test,y_test)
model.fit(x_train,y_train)
x_trainpre=model.predict(x_train)
x_testpre=model.predict(x_test)
train_acc=accuracy_score(x_trainpre,y_train)
testacc=accuracy_score(x_testpre,y_test)
print("training accuracy")
print(train_acc)
print('testing accuracy')
print(testacc)
l=["1.preganancies","2.Glucose","3.BP","4.skinthickness","5.insulin","6.BMI", "7.diabetespedigreefunction","8.age"]
ind=list()
for i in l:
  print(i)
  a=int(input())
  ind.append(a)
ind=np.array(ind).reshape(1,-1)
sc = StandardScaler()
sc.fit(ind)
sd=sc.transform(ind)
pre=model.predict(sd)
if pre==0:
  print("Don t worry you have not diabatics")
else:
  print("you have diabetes take further treatment,you will cure soon")
```
